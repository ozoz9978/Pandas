# Series (시리즈)

## 시리즈 생성
- 시리즈는 1차원 배열과 유사하게 사용할 수 있다.
- Series 클래스의 생성자에 리스트를 전달하여 시리즈 객체를 생성할 수 있다.
- 리스트의 요소에 대해 인덱스가 자동으로 생성되어 시리즈 객체를 출력할 때 표시된다.

```python
import pandas as pd
sr = pd.Series ([2,3,5,7,11])
print(type(sr))
print(sr)
```

```python
# 명시적으로 인덱스 전달하기
menu_list = ['토스트', '제육볶음','치킨']
index_list = ['아침', '점심', '저녁']
sr = pd.Series(menu_list,index_list)
print(sr)
```

```python
# 인덱스와 값 추출
print(sr.index)
print(sr.values)
print(type(sr.values))
```

```python
# 파이썬 딕셔너리를 이용한 Series 생성
dic_data = {'a':21,'b':22,'c':23,'d':24,'e':25}
sr = pd.Series(dic_data, dtype='float32')
```

## 원소 선택

```python

import pandas as pd
print(sr)


print('-'*30)
print(sr[0]) # sr.iloc[0] 동일
print('-'*30)
print(sr['b'])
print('-'*30)
print(sr[:2])
print('-'*30)
print(sr[[0,2]])
print(sr['a':'d']) # 인덱스 이름을 이용한 슬라이싱은 마지막 인덱스도 포함한다.
```

## unique(), value_counts()
- sr.unique(): `결측치를 포함하며` 중복을 제외한 데이터의 종류를 ndarray로 반환
- sr.value_counts : `결측치를 포함하지 않으며` 데이터의 종류별 개수를 Series로 반환

```python
import numpy as np
sr = pd.Series(['A','B','A', np.NaN,'C','D','D','A'])
print(sr.unique())
print(sr.value_counts())
```

# DataFrame(데이터 프레임)

## 데이터 프레임 생성
- 데이터 프레임을 2차원 배열과 유사하게 사용할 수 있다.
- pandas의 dataframe 클래스의 생성자에 이중 리스트를 전달하여 데이터 프레임객체를 생성할 수 있다.
- 리스트의 요소에 대해 인덱스와 컬럼이 자동으로 생성되어 데이터프레임 객체를 출력할 때 표시된다.

```python
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]])
display(df)
print(type(df))
```

```python
# 인덱스와 컬럼을 명시적으로 전달하기
import pandas as pd
menu_list = [['토스트', '시리얼', '스크램블'],
             ['제육볶음', '칼국수', '육개장'],
             ['치킨','삼겹살','라면']]

index = ['아침', '점심', '저녁']  # 저녁 메뉴가 3개이므로 '저녁'을 3번 넣어줍니다.
columns = ['월', '화', '수']

df = pd.DataFrame(menu_list, index=index, columns=columns)
display(df)
```

```python
# 인덱스, 컬럼, 값 가져오기
print(df.index)
print(df.columns)
print(df.values)
```

```python
# 딕셔너리를 이용한 데이터프레임 생성
raw_data = {
    '이름':['홍길동','전우치','손오공','사오정','저팔계'],
    '나이':[32,27,30,31,33],
    '전화번호':['010-1111-1111','010-2222-2222','010-3333-3333','010-4444-4444','010-5555-5555'],
    '지역':['서울','경기','대구','광주','제주']
}
df = pd.DataFrame(raw_data)
display(df)
```

# 데이터 정렬

```python
import numpy as np
import pandas as pd
np.random.seed(10)
data = np.random.randint(100,size = (10,10))
df = pd.DataFrame(data,index=list('ABCDEFGHIJ'),columns=list('abcdefghij'))
print(df)
```

## df.sort_values()
- 행 또는 열을 하나 지정해서 값에 따라 정렬한다.
- 기본 값은 열을 지정해서 값에 따라 오름차순으로 정렬하는 것이다
- 내림차순으로 정렬하려면 ascending 인수에 False 값을 전달한다.

```python
print(df.sort_values('a'))
```

- 행을 지정해서 정렬하려면 axis인수에 1또는 'columns' 값을 전달한다.

```python
print(df.sort_values('A',axis=1))
```

## df.sort_index ()
- 행 또는 열에 대해 정렬한다.
- 기본값은 인덱스를 오름차순으로 정렬하는 것이다
- 컬럼명에 따라 내림차순으로 정렬하려면 axis 인수와 ascending 인수를 설정한다.

```python
print(df.sort_index(ascending=False))
```

```python
print(df.sort_index(axis='columns',ascending=False))
```

# 데이터프레임 조작

## 새로운 행 또는 열 추가
- DataFrame에 새로운 열 추가
      - df[열이름] = 데이터 목록
      - 새로 추가되는 데이터 목록은 기존 데이터의 행(레코드) 수와 같아야 한다.
- DataFrame에 새로운 행 추가
      - df.loc [행이름] = 데이터 목록

```python
import pandas as pd

df = pd.DataFrame([[1, 2, 3], [7, 8, 9], [13, 14, 15]], index=list('ABC'), columns=list('abc'))
print(df)
```

```python
# 새로운 행 추가
df.loc['D'] = [19,20,21]
print(df)
```

```python
# 새로운 열 추가
df['d'] = [4,10,16,22]
print(df)
```

- 시리즈를 이용한 새로운 행 또는 열 추가 `(한 줄씩 추가할때)`
    - 시리즈를 이용하여 새로운 행 또는 열을 추가할 때는 기존 데이터프레임과 같은 `인덱스를 갖도록` 설정한다
    - 인덱스를 명확하게 명시하지 않으면 추가되는 요소는 NaN으로 초기화 된다.

```python
df.loc['E']=pd.Series([25,26,27,28],index=df.columns)
print(df)
```

```python
df['e'] = pd.Series([5,11,17,23,29],index=df.index)
print(df)
```

## df.reindex()
- 인덱스 구조를 변경하면서 새로운 행 또는 열을 추가한다.
- 추가되는 행 또는 열의 요소는 NaN으로 초기화된다.

## pd.concat()
- 두 개 이상의 데이터프레임을 행 또는 열 방향으로 연결한다.
- axis = 0 값이 기본 값으로 행 방향으로 연결한다.
- 열 방향으로 연결하고자 할 경우 axis 인자를 전달한다.
- 행 방향으로 연결하고자 할 때는 열 이름이 같아야 하고 열 방향으로 연결하고자 할 경우에는 행 이름이 같아야 한다.
- 실행결과는 원본에 영향을 주지 않고 새로운 데이터프레임 객체를 반환한다

```python
df2 = pd.concat([df,df])
print(df2)
```

```python
df3 = pd.DataFrame([[5,6],[7,8]],index=df.index,columns=list('cd'))
print(df3)
```

```python
df4 = pd.concat([df,df3],axis=1)
print(df4)
```

```python
import pandas as pd

index_list = ['홍길동', '임꺽정', '전우치', '손오공', '저팔계', '사오정']

sr_major = pd.Series(['컴퓨터공학과', '수학과', '정보통신학과', '수학과', '컴퓨터공학과', '컴퓨터공학과'], index=index_list)
sr_mat = pd.Series([97, 88, 91, 76, 88, 87], index=index_list)
sr_kor = pd.Series([88, 89, 85, 90, 88, 77], index=index_list)
sr_eng = pd.Series([90, 100,96,91, 80, 90], index=index_list)

df = pd.DataFrame({'전공': sr_major, '수학': sr_mat, '국어': sr_kor, '영어': sr_eng})

# '합계' 열 추가
df['합계'] = df.loc[:, '수학':'영어'].sum(axis=1)

# '평균' 열 추가
df['평균'] = df.loc[:, '수학':'영어'].mean(axis=1)

# '평균' 열을 소수점 둘째 자리까지 문자열로 변환
df['평균'] = df['평균'].apply(lambda x: f'{x:.2f}')
df['평균'] = df['평균'].astype(float)
df
```

# 데이터프레임 재구조화

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.arange(16).reshape((4,4)),index=list('ABCD'),columns=list('abcd'))
year_df = pd.DataFrame([2021,2022,2021,2022],index=list('ABCD'),columns=['year'])
class_df = pd.DataFrame(list('AABB'),index=list('ABCD'),columns=['class'])

df=pd.concat([year_df,class_df,df],axis=1)
print(df)
```

```python
## pd.pivot_table() (=df.pivot_table())
- 지정된 index와 column으로 사용하는 피벗 테이블을 생성한다.
- 피벗테이블 : 기존 데이터를 기반으로 합계나 평균등의 통계를 산출하기 위해 새로운 구조의 표를 만드는 기능
- 데이터에서 특정 컬럼 항목으로만 정렬학 싶을 때 피벗 테이블을 사용한다. 그렇게 되면 해당 컬럼이 index가 되고 중복된 컬럼의 항목은 하나로 합쳐지고 값들의 평균을 가지게 된다.
- df.pivot_table(index=행방향컬럼,columns=열방향컬럼, values=집계대상컬럼, aggfunc=구할 통계값)
```

```python
pivot = pd.pivot_table(df,index='year',columns='class')
print(pivot)
```

```python
pivot = pd.pivot_table(df,index=['year','class'])
print(pivot)
```

```python
# 연도를 기준으로 각 컬럼의 평균 값
# aggfunc = np.mean 기본값으로 설정되어 있음
pivot = pd.pivot_table(df,index='year',values=['a','b','c','d'])
print(pivot)
```

```python
# 연도를 기준으로 각 컬럼의 합계 계산
pivot = pd.pivot_table(df,index='year',values=['a','b','c','d'],aggfunc=np.sum)
print(pivot)
```

```python
pivot = pd.pivot_table(df,index='year',values=['a','b','c','d'],aggfunc=['sum','mean'])
print(pivot)
```

```python
[실습] 피벗 테이블 실습
- 'benefit' 컬럼추가: 매출('sales') - 비용('cost')
- 피봇을 활용하여 연도별 분기 이익 데이터 확인
- 연도별 이익의 평균/합계 계산
```

```python
sr_year = pd.Series([2020]*4+[2021]*4+[2022]*4)
sr_quarter = pd.Series(['10','20','30','40']*3)
np.random.seed(0)
sr_sales = pd.Series(np.random.randint(500,6000,12))
sr_cost = pd.Series(np.random.randint(100,1200,12))

df=pd.DataFrame({'year':sr_year,'quarter':sr_quarter,'sales':sr_sales,'cost':sr_cost})
print(df)

```

```python
# 'benefit' 컬럼추가: 매출('sales') - 비용('cost')
df['benefit'] = df['sales']-df['cost']
print(df)
# - 피봇을 활용하여 연도별 분기 이익 데이터 확인
pivot = pd.pivot_table(df,index='year',columns='quarter',values='benefit')
# pivot = pd.pivot_table(df,index=['year','quarter'],values=['benefit'])
print(pivot)

# - 연도별 이익의 평균/합계 계산
pivot = pd.pivot_table(df,index='year',values='benefit',aggfunc=['sum','mean'])
print(pivot)
```

```python
currency_format = df.copy()
for col in ['sales','cost','benefit']:
    currency_format[col] = currency_format[col].map(lambda x:f'￦{x:,}')
print(currency_format)
```

## pd.date_range() 함수
- start: 시작날짜
- period: 생성할 timestamp 개수
- freq: 시간 간격 (D: 1일 default, nD: n일 간격, w:1주, M:월말, MS:월초, Q:분기말, QS: 분기초, A:연말, AS: 연초)

```python
import numpy as np
import pandas as pd

dates = pd.date_range('20240415', periods=6)
print(type(dates))
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
print(df)
```

```python
# 다가오는 가장 이른 월초 값을 시작값으로 갖는다.
dates = pd.date_range('20240415', periods=6,freq='MS')
print(type(dates))
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
print(df)
```

## to_datetime()의 format
- format 속성은 지정한 포맷으로 변경하라는 의미가 아닌, 원본데이터를 주어진 포멧에 맞춰 해석하여 변경하라는 의미
- format 사이트 참조 https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

```python
values = {'dates':['050320243000','160320243000','20032024200000'],
          'status':['Opened','Opened','Closed']}
df = pd.DataFrame(values)

# 기본 날짜 형식이 아니어서 에러발생 df['dates'] = pd.to_datetime(df['dates'])

df['dates'] = pd.to_datetime(df['dates'],format='%d%m%Y%H%M%S')
print(df)
```

```python
## ts.to_period()
- DatetimeIndex 객체를 PeriodIndex 객체로 변환
- freq: 변경하고자 하는 시간 간격
```

```python
dates = ['2023-01-01', '2023-02-01', '2023-03-01']
ts_date = pd.to_datetime(dates)
print(ts_date)
```

```python
# 위 데이터를 일 간격으로 변환
ps = ts_date.to_period(freq='D')
print(ps)
```

```python
# 주 간격으로 변환
# 원본 날짜를 포함하는 월~일까지의 주 간격 날짜 데이터 생성
ps = ts_date.to_period(freq='W')
print(ps)
```

```python
# 월 간격으로 변환
ps = ts_date.to_period(freq='M')
print(ps)
```

# DataFrame 합치기

## pd.merge() 함수
- DataFrame이 병합되기 위해서는 각 데이터프레임에 공통칼럼이 필요하다.

```python
df1 = pd.DataFrame(
    {'번호':[10,20,30,40,50,60,70],
    '이름':['홍길동','임꺽정','전우치','손오공','저팔계','사오정','현수법사']}
    )
print(df1)
```

```python
df2 = pd.DataFrame(
    {'번호':[10,10,20,30,40,50,60,70,80],
    '금액':[1000,2500,3000,4500,2500,7000,2342,1234,9000]}
    )
print(df2)
```

```python
# '번호'컬럼을 공통 키르 사용하여 두 데이터 프레임을 병합
# inner join 양 데이터프레임에 공통으로 있는 키에 한해서 병합 (번호 80은 df2에는 없어서 없음)
print(pd.merge(df1,df2))
```

```python
# outer join
print(pd.merge(df1,df2,how='outer'))
```

```python
# 왼쪽 데이터프레임에 있는 열 값 기준으로 병합

print(pd.merge(df1, df2, how='left')) # df2의 
```

```python
print(pd.merge(df1,df2,how='right'))
```

## join on
- 병합할 데이터프레임에 같은 이름을 갖는 컬럼이 없거나 반대로 동일 컬럼명이 여러개일 경우 병합에 사용할 공통기를 명시적으로 기재하는 방식

```python
df1 = pd.DataFrame(
    {'번호':[10,20,30,40,50,60,70],
    '이름':['홍길동','임꺽정','전우치','손오공','저팔계','사오정','현수법사']}
    )
print(df1)

df2 = pd.DataFrame({
    '고객번호':[10,10,20,40,50,40,40,70,80],
    '금액':[1000,2500,3000,4500,2500,7000,2342,1234,9000]}
    )
print(df2)
```

```python
# 공통 컬럼명이 없는 경우 오류발생
print(pd.merge(df1, df2, left_on='번호',right_on='고객번호'))
```
